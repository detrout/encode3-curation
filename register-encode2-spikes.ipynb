# Create Document

```python
from htsworkflow.submission.encoded import Document
```

```python
encode2spikes_uuid='057a89e2-57d2-417b-9d9e-1292e7bff043'
encode2spikes = Document('Encode2TranscriptsSpikedIn.pdf', 
                 'spike-in concentrations', 
                 'Caltech Transcripts Spiked in during ENCODE2')
pprint.pprint(encode2spikes.create_if_needed(server, encode2spikes_uuid))
```

# Create Dataset

```python
dataset = {
    '@id': '/datasets/ENCSR714LOF/',
    'award': award,
    'lab': lab,
    'description': 'Caltech spike ins used for ENCODE2',
    'aliases': ['barbara-wold:encode2-spikes'],
    'dataset_type': 'spike-ins',
    'documents': [ encode2spikes_uuid ],
}
```

```python
server.validate(dataset, '/datasets/')
```

```python
if '@id' not in dataset:
    pprint.pprint(server.post_json('/datasets/', dataset))
else:
    pprint.pprint(server.get_json(dataset['@id']))
```

# Create File

```python
fasta_id = '/files/ENCFF862GMF/'
fasta = {
    'dataset': dataset['@id'],
    'file_format': 'fasta',
    'md5sum': hashlib.md5(open('encode2spikes.fa', 'rb').read()).hexdigest(),
    'output_type': 'spike-in sequence',
    'award': award,
    'lab': lab,
}
```

```python
server.validate(fasta, '/file/')
```

```python
if fasta_id is None:
    fasta = server.post_json('/file/', fasta)
    fasta = fasta['@graph'][0]
else:
    fasta = server.get_json(fasta_id)
pprint.pprint(fasta)
```
