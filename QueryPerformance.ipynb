# RDFLib

```python
from rdflib import ConjunctiveGraph
from rdflib import BNode, URIRef, Literal
```

```python
def pyldToRDFLibNode(item):
    """Convert JSON-serialized node to a Soprano Node
    """
    nodetype = item['type']
    value = item['value']
    datatype = item.get('datatype', None)

    if nodetype == 'blank node':
        return BNode()
    elif nodetype == 'IRI':
        return URIRef(value)
    else:
        return Literal(value)
```

```python
storage_dir='rdflib-experiments'
need_to_load = False
if not os.path.exists(storage_dir):
    os.mkdir(storage_dir)
    need_to_load = True

graph = ConjunctiveGraph(store='Sleepycat')
graph.open(storage_dir, create=need_to_load)

if need_to_load:
    for triple in experiments:
        s = pyldToRDFLibNode(triple['subject'])
        p = pyldToRDFLibNode(triple['predicate'])
        o = pyldToRDFLibNode(triple['object'])
        graph.add((s, p, o))
```

```python
def queryRDFLib(query):
    query_times = []
    for i in range(5):
        results = []
        tzero = time.time()
        for row in graph.query(query):
            results.append(row)
        query_times.append(time.time()-tzero)
    pprint(results)
    return query_times
```

```python
timings ['rdflib-select-type'] = queryRDFLib('''select distinct ?o where { ?s a ?o}''')
```

```python
import RDF
```

```python
storage_dir = os.getcwd()
options = "hash-type='bdb',dir='{0}'".format(storage_dir)
storage = RDF.HashStorage('experiments-hash', options=options)
model = RDF.Model(storage)
```

```python
if 1: #not os.path.exists(os.path.join(storage_dir, 'experiments-hash-po2s.db')):
    # need load
    tzero = time.time()
    model.load('file://'+os.path.abspath('experiments.turtle'), 'turtle')
    tfinish = time.time()
    print 'Took: {} s'.format(tfinish-tzero)
```

```python
def queryRedlandType():
    query_times = []
    for i in range(5):
        results = []
        tzero = time.time()
        query = RDF.SPARQLQuery('''select distinct ?o where { ?s a ?o } order by ?o''')
        for row in query.execute(model):
            results.append(row)
        query_times.append(time.time()-tzero)
    pprint(results)
    return query_times

timings ['redland-select-type'] = queryRedlandType()
```

```python
query = RDF.SPARQLQuery('''select distinct ?p where { ?s ?p ?o }''')
for row in query.execute(model):
    line = []
    for k in row:
        line.append(str(row[k]))
    print '\t'.join(line)
```

```python
query = RDF.SPARQLQuery('''
prefix submit: <http://submit.encodedcc.org/profiles/experiment.json#> 
select distinct ?o where { ?s submit:lab.title ?o }
order by ?o''')
for row in query.execute(model):
    line = []
    for k in row:
        line.append(str(row[k]))
    print '\t'.join(line)
```
