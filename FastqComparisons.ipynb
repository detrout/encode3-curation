## Initial setup

```python
from __future__ import print_function
from collections import OrderedDict
from idfastq.fastq_summary import FastqSummary, read_pretty, format_lane_id
import sys
import os
from pprint import pprint
import pandas
import numpy
%pylab inline
```

```python
def load_summaries(filenames):
    summaries = []
    for filename in filenames:
        with open(filename, 'r') as instream:
            summaries.extend(read_pretty(instream))

    print('Total recorded fastqs:', len(summaries))
    return summaries
```

```python
def compute_distances(summaries):
    fastqs = {}
    for fastq in summaries:
        distances = [(fastq.distance(x), x) for x in summaries if x != fastq]
        distances.sort()
        fastqs[fastq] = distances
    return fastqs        
```

```python
def group_summaries(a, b):
    notfound = set()
    found = {}
    confusion = {}
    for afastq in a:
        distances = [(afastq.distance(x), x) for x in b]
        distances.sort()
        if distances[0][0] == 0:
            found[afastq] = distances[0][1]
        elif distances[0][0] == 1:
            notfound.add(afastq)
        else:
            for d in distances:
                if d[0] < 1.0:
                    confusion.setdefault(afastq, []).append(d)
    return notfound, confusion, found
```

```python
def compare_summaries_report(a, aname, b, bname, verbose=False):
    print('{}: {}'.format(aname, len(a)))
    print('{}: {}'.format(bname, len(b)))
    if len(a) < len(b):
        b, a = a, b
        bname, aname = aname, bname
        print('--- swapping parameters, longest first ---')
        
    notfound, confusion, found = group_summaries(a, b)
    print('---- not found: {}----'.format(len(notfound)))
    if verbose:
        for k in sorted(notfound):
            print('  {} {} not found'.format(aname, os.path.split(k.filename)[1]))
    print('--- confused: {} ---'.format(len(confusion)))
    if verbose:
        for k in sorted(confusion):
            print('{}: {}'.format(aname, os.path.split(k.filename)[1]))
            for match in confusion[k]:
                print('  {:3} {} {}'.format(match[0], bname, os.path.split(match[1].filename)[1]))
    print('--- found: {} ---'.format(len(found)))
    if verbose:
        for k in sorted(found):
            print('{} {} == {} {}'.format(aname, os.path.split(k.filename)[1], 
                                          bname, os.path.split(found[k].filename)[1]))
      
              
```

## What is related to this file?

Given to groups interate over one, looking to see what files are related to it in the other group.

```python
def relationship(first, second):
    relations = {}
    for fq1 in first:
        for fq2 in second:
            d = fq1.distance(fq2)
            if d < 1:
                relations.setdefault(fq1, {})[fq2] = d
    return relations
    
```

```python
def report_reads(fq, indent='  ', threshold=1e6):
    for r in sorted(fq.reads):
        if fq.reads[r] > threshold:
            print('{}{}: {}'.format(indent, format_lane_id(r), fq.reads[r]))
    
def report_relationship(group):
    for i, fq in enumerate(sorted(group, key=lambda x: os.path.split(x.filename)[1])):
        print('{} {}'.format(i, os.path.split(fq.filename)[1], '->'))
        report_reads(fq, ' '* 2)
        for k in sorted(group[fq], key=lambda x: os.path.split(x.filename[1])):
            print('{}{} ({:.3})'.format(' ' * 4,os.path.split(k.filename)[1], float(group[fq][k])))
            report_reads(k, ' ' * 6)
            
```

```python
def get_read_ids(fastqs, threshold=1e6):
    read_ids = set()
    for fq in fastqs:
        for r in fq.reads:
            if fq.reads[r] > threshold:
                read_ids.add(r)
    return read_ids
```

```python
def build_cm(read_ids, fastqs):
    cm = OrderedDict()
    for fq in fastqs:
        for r in sorted(read_ids):
            if r in fq.reads:
                short_name = short_md5_index[fq]
                cm.setdefault(short_name, {})[format_lane_id(r)] = fq.reads[r]
    confusion_matrix = pandas.DataFrame(cm)
    return confusion_matrix
```

```python
def get_group_name(filename):
    groups = {'initial': initial,
              'august': august,
              'september': september,
              'dcc': dcc}
    for gname in groups:
        group_filenames = [x.filename for x in groups[gname]]
        if filename in group_filenames:
            #return "{} {}".format(gname, os.path.split(filename)[1])
            return (gname, filename)
    print(filename, 'not found')
```

Try to cluster fastq files together into groups that had something in common.

```python
def find_clusters(all_fastqs):
    clusters = []
    for fq in all_fastqs:
        added = False
        for cluster in clusters:
            for element in cluster:
                if fq.distance(element) < 1:
                    cluster.add(fq)
                    added = True
                    break
        if not added:
            clusters.append(set((fq,)))
                
    return clusters
```
