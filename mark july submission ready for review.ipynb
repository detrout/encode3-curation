Approve experiments

we need to approve the experiments/libraries that were uploaded to the DCC

```python
import os
import pandas
```

```python
from curation_common import *
```

```python
server = ENCODED('www.encodeproject.org')
server.load_netrc()
```

```python
submission = pandas.read_excel('/home/diane/Jun2015-encode-submission.xlsx')
```

```python
def set_to_ready(submission, experiment_column, dry_run=True):
    experiments = set()
    for i in submission.index:
        row = submission.irow(i)
        experiment = row[experiment_column]
        if experiment.endswith('/'):
            experiment = experiment[:-1]
        _, experiment = os.path.split(experiment)
        experiments.add(experiment)
    
    changes = 0
    for experiment_id in experiments:
        url = '/experiments/{}/'.format(experiment_id)
        experiment_obj = server.get_json(url)
        status = experiment_obj['status']
        if status in ['submitted', 'started']:
            release = {'status': 'ready for review'}
            changes += 1
            if not dry_run:
                print(server.patch_json(url, release))
    print("changed {} records".format(changes))
```

```python
set_to_ready(submission, 'Dcc exp', dry_run=False)
```
