Find our library ids for non-compliant encode experiments.

so we can try and find whats missing.

```python
from __future__ import print_function
```

```python
from curation_common import *
```

```python
server = ENCODED('www.encodedcc.org')
server.load_netrc()
```

```python
things = server.get_json('https://www.encodeproject.org/search/?type=experiment&lab.title=Barbara+Wold%2C+Caltech&status=released&audit.category=missing%20spikeins_used&award.rfa=ENCODE3')
```

```python
model = get_model()
```

```python
for experiment in things['@graph']:
    print(experiment['@id'])
    load_jsonld_into_model(model, server.get_jsonld(experiment['@id']))
```

```python
%%sparql -m model
select distinct ?experiment ?s ?submitted ?status
where { 
  ?s a file:file ;
     file:dataset ?experiment ;
     file:submitted_file_name ?submitted ;
     file:status ?status .
     filter(regex("released", ?status, "i"))
}
```

```python
%%sparql -m model -o exp_to_library
select distinct ?encode_experiment ?description ?encode_library ?library_id
where { 
  ?rep a replicate:replicate ;
     replicate:library  ?encode_library ;
     replicate:experiment ?encode_experiment .
  ?encode_experiment rdf:description ?description .
  OPTIONAL { ?encode_library library:aliases ?library_id . }
}
order by ?library_id
```

```python
results = []
for row in exp_to_library:
    results.append( [ str(row[c]) for c in exp_to_library.columns ] )
```
