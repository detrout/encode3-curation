```python
%%sql
with Fastq as (
    select uri as FastqUri, 
           payload->'submitted_file_name' as filename,
           payload->'replicate' as replicate,
           -- jsonb_array_elements_text(payload->'flowcell_details') as flowcell_details,
           jsonb_array_length(payload->'flowcell_details') as detail_len
    from item
    where object_type = 'File' and 
          payload->'lab' = '"/labs/barbara-wold/"' and
          payload->'file_type' = '"fastq"'
    ),
    Experiment as (
    select uri as ExperimentUri, 
           jsonb_array_elements_text(payload->'files') as file 
    from item
    where object_type = 'Experiment' and 
          payload->'lab' = '"/labs/barbara-wold/"' 
    )
    select ExperimentUri, FastqUri, filename, detail_len 
    from Fastq join Experiment on FastqUri = Experiment.file
    where detail_len = 0;
```

```python
%%sql
with
  experiment as (
  select uri as Experiment,
         payload->>'accession' as Experiment_Accession,
         payload->>'description' as Experiment_Description,
         payload->>'status' as Experiment_Status,
         payload->>'date_released' as Experiment_Released,
         payload->>'assay_title' as Experiment_Type,
         jsonb_array_elements_text(payload->'replicates') as Replicate
  from item
  where object_type = 'Experiment' and
        payload->'lab' @> '"/labs/barbara-wold/"'::jsonb
  ),
  replicate as (
    select uri as Replicate,
           payload->>'library' as Library
    from item
    
    where object_type = 'Replicate'
  ),
  library as (
    select uri as Library,
           payload->>'accession' as Library_Accession,
           payload->>'date_created' as Library_Created,
           payload->>'biosample' as Biosample,
           substring(jsonb_array_elements_text(payload->'aliases') from 'barbara-wold:([0-9]+)')
               as Jumpgate_Library_ID
    from item
    where object_type = 'Library'
  ),
  Fastq as (
    select uri as FastqUri, 
           payload->>'submitted_file_name' as filename,
           payload->>'replicate' as replicate,    
           -- jsonb_array_elements_text(payload->'flowcell_details') as flowcell_details,
           jsonb_array_length(payload->'flowcell_details') as detail_len
    from item
    where object_type = 'File' and 
          payload->'file_type' = '"fastq"'
  ),
  FastqCount as (
    select count(uri) as fastq_count,
           payload->>'replicate' as replicate
    from item
    where object_type = 'File' and
           payload->>'file_type'='fastq'
    group by replicate
  )
    
select  Experiment_Accession,
--                Experiment_Status,
                Library_Accession,
                to_char(library.Library_Created::date, 'YYYY-MM-DD') as Library_Created,
                fastq_count,
                filename,
                Jumpgate_Library_ID,
                experiments_flowcell.flowcell_id
from experiment
     LEFT JOIN replicate ON experiment.Replicate = replicate.Replicate
     LEFT JOIN library on replicate.Library = library.Library
     LEFT JOIN Fastq on Fastq.replicate = replicate.Replicate
     LEFT JOIN FastqCount on FastqCount.replicate = replicate.Replicate
     LEFT JOIN samples_library on Jumpgate_Library_ID = samples_library.id
     LEFT JOIN experiments_lane on samples_library.id = experiments_lane.library_id
     LEFT JOIN experiments_flowcell on experiments_flowcell.id = experiments_lane.flowcell_id
where fastq_count = 1 and
      filename LIKE '%_l1.fastq%' and
      experiments_flowcell.flowcell_id LIKE '%BCXX'
order by experiment_accession
limit 300
;
```
